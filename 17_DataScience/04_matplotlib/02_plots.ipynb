###  Scatter Plot

```python
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Bar Chart

```python
plt.bar(['A', 'B', 'C'], [3, 7, 5])
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Horizontal Bar Chart

```python
plt.barh(['A', 'B', 'C'], [3, 7, 5])
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

### Pie Chart

```python
plt.pie([10, 20, 30], labels=['A', 'B', 'C'], autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

### Histogram

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Box Plot

```python
data = [np.random.randn(100) for _ in range(4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Area Plot

```python
x = np.arange(1, 6)
y = np.array([1, 2, 3, 4, 5])

plt.fill_between(x, y)
plt.title("Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Error Bar Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
error = 0.1
plt.errorbar(x, y, yerr=error, fmt='o')
plt.title("Error Bar Plot")
plt.show()
```

### Multiple Lines

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title("Multiple Lines")
plt.legend()
plt.show()
```

### Subplots

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].scatter([1, 2, 3], [10, 20, 30])
axs[1, 0].bar(['A', 'B', 'C'], [3, 7, 5])
axs[1, 1].pie([10, 20, 30], labels=['A', 'B', 'C'])
plt.show()
```

### Heatmap

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title("Heatmap")
plt.colorbar()
plt.show()
```

### Contour Plot

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contour(X, Y, Z)
plt.title("Contour Plot")
plt.show()
```

### 3D Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([1, 2, 3], [4, 5, 6], [7, 8, 9])
plt.title("3D Scatter Plot")
plt.show()
```

###  Polar Plot

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(theta))
plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

### Step Plot

```python
plt.step([1, 2, 3], [4, 5, 6])
plt.title("Step Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Stacked Bar Chart

```python
N = 5
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
ind = np.arange(N)
plt.bar(ind, men_means, label='Men')
plt.bar(ind, women_means, bottom=men_means, label='Women')
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

### Donut Chart

```python
sizes = [10, 20, 30]
plt.pie(sizes, labels=['A', 'B', 'C'], startangle=90, pctdistance=0.85)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```

### Annotations

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.annotate('Point (2,5)', xy=(2, 5), xytext=(3, 6), arrowprops=dict(facecolor='black'))
plt.title("Annotations")
plt.show()
```

### Customizing Ticks

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.xticks([1, 2, 3], ['One', 'Two', 'Three'])
plt.title("Customizing Ticks")
plt.show()
```

### Adding Grid

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.grid()
plt.title("Adding Grid")
plt.show()
```

### Background Color

```python
plt.figure(facecolor='lightblue')
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Background Color")
plt.show()
```

### Saving a Figure

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Save Figure Example")
plt.savefig('figure.png')
plt.show()
```

### Logarithmic Scale

```python
x = np.linspace(1, 10, 100)
y = np.exp(x)
plt.plot(x, y)
plt.yscale('log')
plt.title("Logarithmic Scale")
plt.show()
```

### Filled Contour Plot

```python
plt.contourf(X, Y, Z, levels=20)
plt.colorbar()
plt.title("Filled Contour Plot")
plt.show()
```

### Violin Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.violinplot(data)
plt.title("Violin Plot")
plt.show()
```

### Streamplot

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
plt.streamplot(X, Y, U, V)
plt.title("Streamplot")
plt.show()
```
