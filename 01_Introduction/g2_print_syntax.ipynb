### print function 

```python
print("hello", "python", 12132, end="\t", sep=";")
print("world")  # default end='\n'
```

```python
print(1, 2, 3, 4, 5, sep=";", end="\t")
print("a", "b", "c", "d", sep="-")  # default end='\n'
```

```python
# \b - overwrites previous character
print("He\bi")
```

```python
print("12\b34")
```

```python
print("first\bsecond")
```

```python
print("\bsecond")
```

```python
'\bsecond'
```

```python
print("second\b")
```

```python
# \r - overwrites complete existing line
print("He\ri")
print("12\r34")
print("first\rsecond")
print("abcdef\r1234")
print("1234567\rDOG")
```

```python
# Unicode characters \uXXX - unicode character
print("\u20B9")  # ₹   - unicode
print("\u018e")  # Ǝ   - unicode

print("\046")    # &   - Ascii
```

```python
# \x... - hexadecimal number
print("\x24")
print("\xf1")
```
