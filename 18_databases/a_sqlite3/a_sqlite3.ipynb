# sqlite3

```python
import sqlite3
```

```python
help(sqlite3)
```

```python
#  Connecting to a Database
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()

cursor
```

```python
# Creating a Table

cursor.execute('''
CREATE TABLE users (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
''')

connection.commit()
```

```python
# Inserting Data

cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Alice", 25))
connection.commit()
```

```python
# Inserting Multiple Rows

users = [
        ("Bob", 30), 
        ("Charlie", 35)
]
cursor.executemany("INSERT INTO users (name, age) VALUES (?, ?)", users)
connection.commit()
```

```python
# Querying Data


cursor.execute("SELECT * FROM users")
```

```python
cursor.fetchone()
```

```python
cursor.fetchall()
```

```python
cursor.fetchall()  # disposable object
```

```python
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
#  Filtering Data
cursor.execute("SELECT * FROM users WHERE age > ?", (28,))
filtered_rows = cursor.fetchall()
for row in filtered_rows:
    print(row)
```

```python
# Updating Data

cursor.execute("""
                UPDATE users 
                SET age = ? 
               WHERE name = ?""", 
               (26, "Alice"))
connection.commit()
```

```python
#  Deleting Data

cursor.execute("DELETE FROM users WHERE name = ?", ("Bob",))
connection.commit()
```
